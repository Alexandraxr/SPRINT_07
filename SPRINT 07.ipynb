```python
#creo la función
def calcular_imc():
    
        
    peso = float(input("Indica cual es tu peso en KG: "))
    
    altura = float(input("Indica cuanto mides en metros: "))

    # Calcular IMC
    imc = peso / (altura ** 2)


    if imc < 18.5: 
      print("IMC indica : peso bajo")
     
    elif  imc < 25:
      print("IMC indica :  peso normal")
    
    elif imc < 30:
      print("IMC indica :  Sobrepeso")
    
    else:
      print("IMC indica : Obesidad")
      
  
#llamo la función
calcular_imc()

   
```

Exercici 4
Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python
def diccionario_invertido(diccionari_original):  #defino funcion
   
    diccionario_inverso = {}      # Crear un nuevo diccionario para guardar resultado invertido
    
    for clau, valor in diccionari_original.items():
       
        if valor in diccionario_inverso:
            print(f" El valor '{valor}' de la clave '{clau}' ya existe en el diccionario inverso.")
            
        else:
            diccionario_inverso[valor] = clau  # si no esta duplicado, se añade en el diccionario inverso , modificando la clave por el valor
            

    print(f"El diccionario invertido es: {diccionario_inverso}" )          
         
    
diccionario_original = {'x': "apple", 'y':"banana", 'z':"banana", 'e':"tomato", 'l': "carrot", 'h':"apple"}

print(f'El diccionario original es: {diccionario_original}')

diccionario_invertido(diccionario_original)
```

Exercici 1
Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se com una llista. Tingues en compte que si és un valor únic no ha de ser una llista.



```python


def reverse_dict_with_duplicates(diccionario_original):
    reversed_dict = {}


    for key, value in diccionario_original.items(): #Se recorre por cada clave-valor del diccionario original.
        if value in reversed_dict: #Si el valor ya existe en reversed_dict:
            if isinstance(reversed_dict[value], list): #Si el valor en reversed_dict ya es una lista, se agrega la nueva clave a esa lista.
                reversed_dict[value].append(key) #Si el valor en reversed_dict no es una lista, se convierte en una lista que contiene tanto el valor original como la nueva clave.
            else:
                reversed_dict[value] = [reversed_dict[value], key]  #Si el valor no existe en reversed_dict, se añade como una nueva entrada.
                
        else:
            reversed_dict[value] = key

    return reversed_dict

diccionario_original =  {'x': "apple", 'y':"banana", 'z':"banana", 'e':"tomato", 'l': "carrot", 'h':"apple"}

reversed_dict = reverse_dict_with_duplicates(diccionario_original)

print("Lista de valores y claves :", reversed_dict)
```

Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

```python
Numbers_list = [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

def separar_convertibles_no_convertibles(Numbers_list):
    convertibles=[]
    no_convertibles=[]

    def convertible_a_float(x):
       try:
        float(x)
        return True
       except (ValueError,TypeError):
          return False
       
    def crear_lista(lista):
          for x in lista:
             
             if isinstance (x,(list,tuple)):  # si el numero es una lista o tupla lo añadimos en crear_lista, tupla [] lista ()
                crear_lista(x)
            
             else:
                                
                  if convertible_a_float(x):  # si es convertible (funcion) lo añade en lista convertibles 
                    convertibles.append(x)
                  else:
                    no_convertibles.append(x) # si no es convertible lo añade en lista no convertible

    crear_lista(Numbers_list)                         #llama la funcion crear lista, teniendo en cuena la lista : numbers_list 
    return convertibles, no_convertibles
 

convertibles, no_convertibles = separar_convertibles_no_convertibles(Numbers_list)
    
print("Numeros convertibles a flotantes:", convertibles)
print("Numeros no convertibles a flotantes:", no_convertibles)    


#Explicación codigo:
#Error de tipo" Que sale cuando usamos un tipo de dato incorrecto.
#valuerror excepción incorporada (definida por Python) que es lanzada cuando una función espera un valor determinado pero recibe otro.
#Primero, se ejecuta la cláusula try (la(s) linea(s) entre las palabras reservadas try y la except).
#Si no ocurre ninguna excepción, la cláusula except se omite y la ejecución de la cláusula try finaliza.
# si su tipo coincide con la excepción nombrada después de la palabra clave except, se ejecuta la cláusula except, y luego la ejecución continúa después del bloque try/except.
#Al escribir funciones, puedes usar isinstance para asegurarte de que los argumentos pasados son del tipo esperado.
```

Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. Per exemple, per a l'arxiu "tu_me_quieres_blanca.txt" la sortida esperada seria:

****LO HE COMPLETADO PERO NO LO QUIERO ENTREGAR YA QUE TENGO QUE PROFUNDIZARLO.****

```python
import json
import string
from collections import defaultdict

# Funció per comptar i endreçar paraules
def comptador_endreçador_paraules(tu_me_quieres_blanca):
    # Llegeix l'arxiu
    with open(tu_me_quieres_blanca, 'r', encoding='utf-8') as file:
        text = file.read()
    
        # Processa el text: elimina puntuacions i converteix a minúscules
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Divideix el text en paraules
    paraules = text.split()
    
    # Comptador de freqüència de paraules
    freq_paraules = defaultdict(int)
    for paraula in paraules:
        freq_paraules[paraula] += 1
    
    # Organitza les paraules dins d'un diccionari segons la lletra inicial
    diccionari_organitzat = defaultdict(lambda: defaultdict(int))
    for paraula, freq in freq_paraules.items():
        lletra_inicial = paraula[0]
        diccionari_organitzat[lletra_inicial][paraula] = freq
    
    # Ordena les paraules dins de cada lletra inicial
    diccionari_ordenat = {}
    for lletra, paraules_dict in diccionari_organitzat.items():
        diccionari_ordenat[lletra] = dict(sorted(paraules_dict.items()))
    
    # Ordena les lletres inicials
    diccionari_ordenat = dict(sorted(diccionari_ordenat.items()))
    
    return diccionari_ordenat

# Funció per imprimir el diccionari ordenat
def imprimir_diccionari_ordenat(diccionari_ordenat):
    for lletra, paraules_dict in diccionari_ordenat.items():
        print(f"Lletra {lletra.upper()}:")
        for paraula, freq in paraules_dict.items():
            print(f"  {paraula}: {freq}")
        print()

# Exemple d'ús
nom_arxiu = "tu_me_quieres_blanca.txt"  # Substitueix amb el nom del teu arxiu
diccionari_ordenat = comptador_endreçador_paraules(nom_arxiu)
imprimir_diccionari_ordenat(diccionari_ordenat)
```
